```python
import pickle
from sklearn.preprocessing import LabelEncoder


# List of categorical columns in your dataset
categorical_columns = ['Customer type', 'Branch', 'City', 'Product line', 'Payment', 'Gender']

encoders = {}

for col in categorical_columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    encoders[col] = le

# Save the encoders to a pickle file
with open("encoders.pkl", "wb") as f:
    pickle.dump(encoders, f)

print("LabelEncoders saved to encoders.pkl")
```

```python

from sklearn.model_selection import train_test_split
y=df['Gender']
x=df.drop('Gender',axis=1)
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0,test_size=0.2)

```

```python

from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=7)
knn.fit(x_train,y_train)

from sklearn.metrics import accuracy_score,classification_report,confusion_matrix


y_pred=knn.predict(x_test)
print("Classification Report is:\n",classification_report(y_test,y_pred))
print("Confusion Matrix:\n",confusion_matrix(y_test,y_pred))
print("Training Score:\n",knn.score(x_train,y_train)*100)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier(max_depth=6, random_state=123,criterion='entropy')
dtree.fit(x_train,y_train)

from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
from sklearn.metrics import r2_score

y_pred=dtree.predict(x_test)
print("Classification Report is:\n",classification_report(y_test,y_pred))
print("Confusion Matrix:\n",confusion_matrix(y_test,y_pred))
print("Training Score:\n",dtree.score(x_train,y_train)*100)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier()
rfc.fit(x_train,y_train)

from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
from sklearn.metrics import r2_score
y_pred=rfc.predict(x_test)
print("Classification Report is:\n",classification_report(y_test,y_pred))
print("Confusion Matrix:\n",confusion_matrix(y_test,y_pred))
print("Training Score:\n",rfc.score(x_train,y_train)*100)
```

```python

from sklearn.ensemble import GradientBoostingClassifier
gbc=GradientBoostingClassifier()
gbc.fit(x_train,y_train)

from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
from sklearn.metrics import r2_score
y_pred=gbc.predict(x_test)
print("Classification Report is:\n",classification_report(y_test,y_pred))
print("Confusion Matrix:\n",confusion_matrix(y_test,y_pred))
print("Training Score:\n",gbc.score(x_train,y_train)*100)
```

```python



your_model = gbc  # Example: Using GradientBoostingClassifier

# Get predictions
predictions = your_model.predict(x_test)

# Create a comparison DataFrame
comparison_df = pd.DataFrame({'Actual': y_test, 'Predicted': predictions})

# Display the DataFrame or iterate through values
print(comparison_df.head())  # Display first few rows

# Alternatively, iterate and print
for actual, predicted in zip(y_test, predictions):
    print(f"Actual: {actual}, Predicted: {predicted}")
```

```python
import pickle

# Choose a filename for your model
filename = 'gbc_model.pkl'

# Save the model to the file
with open(filename, 'wb') as file:
    pickle.dump(gbc, file)

print(f"Model saved as {filename}")
```
